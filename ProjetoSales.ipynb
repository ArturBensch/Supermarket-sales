# **Análises de *Produtos* e *Vendas***  







```python
vendas_por_produto = tabela_sales.groupby('Product line')['Total'].sum()
vendas_por_produto_USD = vendas_por_produto.map(lambda x: f"${x:,.2f} USD")


##-----------------------Gráfico--------------##


plt.figure(figsize=(10,6))
vendas_por_produto.plot(kind='bar')
plt.title('Vendas por Produto')
plt.xlabel('Produto')
plt.ylabel('Vendas (USD)')

for index, value in enumerate(vendas_por_produto):
  plt.text(index, value, vendas_por_produto_USD.iloc[index], ha='center', va='bottom')

plt.show()
```

```python
analise_produto = tabela_sales.groupby('Product line').agg(
    Total_Vendas=('Total', 'sum'),
    Quantidade_Vendida=('Quantity', 'sum')).reset_index()



##---------------------------Gráfico------------------------##
## Gráfico de dispersão (Relação entre Total de Vendas e Quantidade Vendida)
plt.figure(figsize=(10,6))
sns.scatterplot(
    x= analise_produto['Quantidade_Vendida'],
    y= analise_produto['Total_Vendas'],
    hue= analise_produto['Product line'],
    palette='viridis',
    s=100)

plt.title('Relação entre Total de Vendas e Quantidade Vendida')
plt.grid(True)

plt.show()
```

```python
lucratividade_produtos = tabela_sales.groupby('Product line')['gross income'].sum()
lucratividade_produtos_USD = lucratividade_produtos.map(lambda x: f"${x:,.2f} USD")
display(lucratividade_produtos_USD)

##----------------------------Gráfico-----------------#

# Criar o gráfico de pizza (rosca)
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    lucratividade_produtos,
    labels=lucratividade_produtos.index,
    autopct='%1.1f%%',  # Exibir porcentagens
    startangle=140,
    colors=plt.cm.Paired.colors,
    wedgeprops={'edgecolor': 'black'}
)

# Adicionar um círculo branco no centro para criar o efeito de rosca
centro_branco = plt.Circle((0, 0), 0.50, color='white')  # Raio ajustável para o círculo
plt.gca().add_artist(centro_branco)

# Adicionar o valor total dentro do círculo branco
total = lucratividade_produtos.sum()
plt.text(0, 0, f'Total\n${total:,.2f}', fontsize=14, fontweight='bold', ha='center', va='center')

# Melhorar a aparência dos textos das fatias
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(10)

plt.title('Lucratividade por Linha de Produto')
plt.axis('equal')  # Mantém o gráfico circular

plt.show()
```

```python
tabela_sales['Date']=pd.to_datetime(tabela_sales['Date'])
tabela_sales['Month']=tabela_sales['Date'].dt.month
tabela_sales['Year']=tabela_sales['Date'].dt.year

sazonalidade_produtos=tabela_sales.groupby(['Year','Month','Product line'])['Total'].sum().reset_index()
sazonalidade_produtos=sazonalidade_produtos.sort_values(by=['Year','Month','Total'],ascending=False)


##-----------------Gráfico-----------------##
plt.figure(figsize=(12,6))
sazonalidade_produtos_pivot=sazonalidade_produtos.pivot_table(index='Product line',columns=['Year','Month'],values='Total',aggfunc='sum')#Caso haja valores duplicados, some.#
sns.heatmap(sazonalidade_produtos_pivot, annot=True, fmt='.1f', cmap='YlGnBu', cbar_kws={'label': 'Total de Vendas (USD)'})
plt.title('Sazonalidade de Vendas por Produto (Heatmap)', fontsize=16)
plt.xlabel('Ano e Produto', fontsize=12)
plt.ylabel('Mês', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout() ##Ajusta automaticamente o layout do gráfico para que todos os elementos (títulos, rótulos, legendas, etc.) se ajustem bem na figura, sem sobreposição.
plt.show()
```

```python
avaliacao_produtos = tabela_sales.groupby('Product line')['Rating'].mean()
display(avaliacao_produtos)

##---------------Gráfico-----------##
##Gráfico de Boxplot para Ver a Distribuição dos Ratings

plt.figure(figsize=(12, 6))
sns.boxplot(x='Product line', y='Rating', data=tabela_sales, palette='coolwarm')
plt.title('Distribuição das Avaliações por Linha de Produto', fontsize=16)
plt.xlabel('Linha de Produto', fontsize=12)
plt.ylabel('Avaliação', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
vendas_por_filial = tabela_sales.groupby(['Branch', 'Product line'])['Total'].sum().reset_index()


##-----------------Gráfico-----------------##


plt.figure(figsize=(12, 6))
sns.barplot(
    x='Branch',
    y='Total',
    hue='Product line',
    data=vendas_por_filial,
    palette='viridis'
)
plt.title('Total de Vendas por Filial e Linha de Produto')
plt.xlabel('Filial')
plt.ylabel('Total de Vendas (USD)')
plt.legend(title='Linha de Produto')
plt.show()
```

```python
tabela_sales['Day of Week'] = tabela_sales['Date'].dt.day_name()

vendas_por_dia_da_semana = tabela_sales.groupby('Day of Week')['Total'].sum()
vendas_por_dia_da_semana_USD = vendas_por_dia_da_semana.map(lambda x: f"${x:,.2f} USD")
vendas_totais_por_dia = vendas_por_dia_da_semana_USD.sort_values()
display(vendas_totais_por_dia)

##-----------Gráfico de Linha-------------##
plt.figure(figsize=(10, 6))
sns.lineplot(
    x=vendas_totais_por_dia.index,
    y=vendas_por_dia_da_semana.values,
    marker='o',
    color='b',
    linewidth=2
)

for index, value in enumerate(vendas_por_dia_da_semana.values):
    plt.text(index, value, f"${value:,.0f}", ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.title('Tendência de Vendas por Dia da Semana')
plt.xlabel('Dia da Semana')
plt.ylabel('Total de Vendas (USD)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
tabela_sales['Hour'] = pd.to_datetime(tabela_sales['Time']).dt.hour
vendas_por_hora=tabela_sales.groupby('Hour')['Total'].sum()
vendas_por_hora_USD = vendas_por_hora.map(lambda x: f"${x:,.2f} USD")
vendas_totais_por_hora = vendas_por_hora_USD.sort_values(ascending=False)
display(vendas_totais_por_hora)

##------...----##

vendas_por_cliente_hora = tabela_sales.groupby(['Customer type', 'Hour'])['Total'].sum().unstack()
display(vendas_por_cliente_hora)

plt.figure(figsize=(10, 6))
sns.heatmap(vendas_por_cliente_hora, cmap='coolwarm', annot=True, fmt='.0f')
plt.title('Densidade de Vendas por Hora e Tipo de Cliente')
plt.xlabel('Hora do Dia')
plt.ylabel('Tipo de Cliente')
plt.show()
```

```python
## Diagrama de Pareto ##


# Agrupar e ordenar as vendas por linha de produto
vendas_produto = tabela_sales.groupby('Product line')['Total'].sum().sort_values(ascending=False)

# Calcular a porcentagem acumulada
porcentagem_acumulada = vendas_produto.cumsum() / vendas_produto.sum() * 100

# Criar o gráfico de Pareto
fig, ax1 = plt.subplots(figsize=(12, 6))

# Gráfico de barras para valores absolutos
vendas_produto.plot(kind='bar', color='skyblue', ax=ax1)
ax1.set_ylabel('Total de Vendas (USD)', fontsize=12)
ax1.set_xlabel('Linha de Produto', fontsize=12)
ax1.set_title('Análise de Pareto - Receita por Linha de Produto', fontsize=16)

# Adicionar linha de porcentagem acumulada
ax2 = ax1.twinx()
ax2.plot(vendas_produto.index, porcentagem_acumulada, color='red', marker='o', linewidth=2, label='Porcentagem Acumulada')
ax2.set_ylabel('Porcentagem Acumulada (%)', fontsize=12)
ax2.set_ylim([0, 110])  # Garantir que a escala vá até 100%

# Adicionar linha de referência para 80% (regra de Pareto)
ax2.axhline(y=80, color='green', linestyle='--', label='80% do Total')

# Adicionar legendas e exibir o gráfico
ax2.legend(loc='lower right')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
